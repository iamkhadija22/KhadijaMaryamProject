```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, ttest_1samp
```

```python
#read the csv file
df = pd.read_csv("diabetes_risk_dataset.csv")
```

```python
print("Descriptive Statistics:")
print(df["diabetes_risk_score"].describe())
```

```python
sample_random = df.sample(n=150, random_state=42)

print("Descriptive Statistics for Random Sample:")
print(sample_random["diabetes_risk_score"].describe())
```

```python
sample_systematic = df[df["bmi"] >= 30]

print("Systematic Sample Size:", sample_systematic.shape[0])
print(sample_systematic["diabetes_risk_score"].describe())
```

```python
desc_report = df["diabetes_risk_score"].describe()
desc_report
```

```python
plt.scatter(df["bmi"], df["diabetes_risk_score"], alpha=0.5)
plt.title("Scatter Plot: BMI vs Diabetes Risk Score")
plt.xlabel("BMI")
plt.ylabel("Diabetes Risk Score")
plt.show()
```

```python
sns.boxplot(x=df["diabetes_risk_score"])
plt.title("Box Plot: Diabetes Risk Score")
plt.show()
```

```python
plt.hist(df["diabetes_risk_score"], bins=20)
plt.title("Histogram: Diabetes Risk Score")
plt.xlabel("Diabetes Risk Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include="number").corr())
plt.title("Heat Map: Correlation")
plt.show()
```

```python
r, p_value = pearsonr(df["bmi"], df["diabetes_risk_score"])

print("Pearson Correlation:")
print("p-value =", p_value)
```

```python
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to reject H0")
```

```python

sample = df["diabetes_risk_score"].sample(150)

t_stat, p_val = ttest_1samp(sample, df["diabetes_risk_score"].mean())

print("One Sample T-Test:")
print("p-value =", p_val)
```

```python
if p_val < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to reject H0")
```
